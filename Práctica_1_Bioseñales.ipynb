1) Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva
una serie con la nota mínima, la máxima, media y la desviación típica.

```python
import numpy as np

def estadisticas_notas(diccionario):

    notas = list(diccionario.values())

    minimo = min(notas)
    maximo = max(notas)
    media = np.mean(notas)
    desviacion = np.std(notas)

    return minimo, maximo, media, desviacion
```

```python
notas = {
    "Ana": 4.5,
    "Luis": 3.8,
    "Carlos": 4.2,
    "Sofia": 4.9
}

resultado = estadisticas_notas(notas)

print("Mínima:", resultado[0])
print("Máxima:", resultado[1])
print("Media:", resultado[2])
print("Desviación típica:", resultado[3])
```

**2.** Cálculo del Índice de Masa Corporal (IMC) con Pandas.

```python
import pandas as pd
import numpy as np
```

```python
# a) carga de datos de csv
df = pd.read_csv('pacientes_VitalDB (1).csv')
#b) Mostrar primeras y últimas filas
print(df.head())
print(df.tail())
```

```python
#Renombramiendo de las columnas weight y height por poca coincidencia de los datos del df original
df['weight'], df['height'] = df['height'], df['weight']
print(df.head())
```

```python
# c) Descripción general
df.describe()
```

```python
# d) Calcular IMC
def calc_imc(peso,altura):
  # Convertimos altura de cm a metros
  conv_alt = altura/100
  # Fórmula IMC
  imc = peso/(conv_alt**2)
  return imc

for i in df:
  df['BMI']= calc_imc(df['weight'],df['height'])
# e) Mostrar resultado con la columna BMI agregada
print(df.head())
```

```python
# f) Categorización por IMC
def categoria_imc(bmi):
    if bmi < 18.5:
        return "Bajo peso"
    elif 18.5 <= bmi <= 24.9:
        return "Normal"
    elif 25 <= bmi <= 29.9:
        return "Sobrepeso"
    else:
        return "Obesidad"

df['Categoria_IMC'] = df['BMI'].apply(categoria_imc)

print(df.head())
```
